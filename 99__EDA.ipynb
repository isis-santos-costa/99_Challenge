üîé <i> By searching for correlations, we can understand pairwise interactions between variables and also check for possible reductions towards <b> data efficiency</b>.

### Predictor x Predictor: Trip duration <i>vs.</i> distance

```python
# Checking correlation trip duration x distance
x, y = df['duration_minutes'], df['trip_distance']

plt.figure(figsize=(20, 6))
plt.scatter(x, y, c='blue')
plt.title('Checking correlation'+ y.name + ' x '+ x.name)
plt.xlabel(x.name)
plt.ylabel(y.name)
plt.show()
```

<i> Those two points below 30 minutes, close to 4,000 miles and 34,000 miles, respectively, correspond to around <b>8k miles/hour</b> and <b>~70k miles/hour</b>.
<p> <i>Even for some extravagant scientific experiment of maybe Tesla visionaires this seem a little too much: safe to drop. </i>üßπ

```python
indexNames = df[ (df['trip_distance'] > 2000) ].index
df.drop(indexNames, inplace=True)
max(df['trip_distance'])
```

```python
# Checking correlation trip duration x distance
x, y = df['duration_minutes'], df['trip_distance']

plt.figure(figsize=(20, 6))
plt.scatter(x, y, c='blue')
plt.title('Checking correlation'+ y.name + ' x '+ x.name)
plt.xlabel(x.name)
plt.ylabel(y.name)
plt.show()
```

üßπ <i> 1,600 miles/hour still not credible: let's drop further.</i>

```python
indexNames = df[ (df['trip_distance'] > 800) ].index
df.drop(indexNames, inplace=True)
max(df['trip_distance'])
```

```python
# Checking correlation trip duration x distance
x, y = df['duration_minutes'], df['trip_distance']

plt.figure(figsize=(20, 6))
plt.scatter(x, y, c='blue')
plt.title('Checking correlation'+ y.name + ' x '+ x.name)
plt.xlabel(x.name)
plt.ylabel(y.name)
plt.show()
```
